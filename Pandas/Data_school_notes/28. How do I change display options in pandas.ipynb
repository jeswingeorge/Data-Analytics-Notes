# How do I change display options in pandas

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
#drinks
```

Here when printing all rows of dataframe an elipsis comes after 1st 30 rows and then after hiding some rows shows the remaining 30 rows. Pandas does it so that it does not overrun your display.

What if we want to see the whole rows?

[Pandas doc get_option](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_option.html)


```python
pd.get_option('display.max_rows')    # Pandas displays only 60 rows of dataframe
```

```python
# to change this
pd.set_option('display.max_rows',None)   # to show all 193 rows of dataframe
```

```python
#drinks  # Now we can see all rows of dataframe
```

```python
pd.get_option('display.max_columns')   # to see max number of columns which will be displayed
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

here __Name__ in the 2nd row has elipsis but we want to see the full name or all characters in the cell.

```python
pd.get_option('display.max_colwidth')   # so only 1st 50 characters will be shown in a cell
```

```python
pd.set_option('display.max_colwidth',1000)   # cant use None here
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()    # able to see the whole name
```

Now in __Fare__ column if we dont wanr till 4 decimal pts but only till 3 decimal points.

```python
pd.get_option('display.precision')
```

```python
pd.set_option('display.precision',3)
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()    # Now i fare able to see only 3 decimal points
```
