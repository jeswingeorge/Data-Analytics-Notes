## Dummy variables/ Indicator variables

Used to include unordered categorical variables into machine learning models

```python
import numpy as np
import pandas as pd
pd.__version__
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

Need to make a dummy variable for the sex column.

1. One way to do it is to use - __Series map method__.

Assign - Female - 0  Male - 1

```python
train['sex_male'] = train['Sex'].map({'female' : 0, 'male' : 1})
train.head()
```

2. Other way to do it pd.get_dummies and pass categorical series to it.
  It creates a column for every possible value.
  And puts 1 for each row where value corresponds.
  
  
__If you have k possible values for a categorical variable you will use (k-1) dummy variables to represent it. As we have 2 possible values for sex so we will use only 1 dummy variable to represent it as it will capture all possible info about the column.__ The dropped column is known as __baseline__. Its more flexible.

So here we dropping the first column here (female) as we dont need it and we define it as the baseline.


```python
pd.get_dummies(train['Sex']).iloc[:,1:].head(11)
```

```python
pd.get_dummies(train['Sex'], prefix='Sex').iloc[:,1:].head(6)
```

### Trying to get dummies for categorical variables having more than 2 values

```python
train['Embarked'].value_counts()
```

```python
pd.get_dummies(train['Embarked'], prefix='Embarked').head(11)
```

As before we need only, k-1 dummy variables. 

```python
pd.get_dummies(train['Embarked'], prefix='Embarked').iloc[:,1:].head(11)
```

To identify which port the variables represent

1,0  - Q  
0,1  - S  
0,0  - C


```python
embarked_dummies = pd.get_dummies(train['Embarked'], prefix='Embarked').iloc[:,1:]
```

### How to attach this to dataframe  - using pd.concat

```python
train = pd.concat([train, embarked_dummies], axis = 1)
train.head()
```

### How to pass a dataframe to pd.get_dummies

```python
pd.get_dummies(train, columns = ['Sex', 'Embarked']).head(7)
```

Observed there is no `Sex` and `Embarked` column.

But now how to drop one column of each group.

There is a new parameter in the pandas,get_dummies() - __drop_first = True__. This drops the first column.

Observe `Sex_female` and `Embarked_C` is dropped.

```python
pd.get_dummies(train, columns = ['Sex', 'Embarked'], drop_first = True).head(7)
```
