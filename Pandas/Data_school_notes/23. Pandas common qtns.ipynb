## How to randomly sample rows in pandas?

```python
ufo.sample(n=3)
```

```python
ufo.sample(n=3)
```

```python
ufo.sample(n=3, random_state = 42)
```

```python
# to gte a fraction of total rows
ufo.sample(frac = 0.0001, random_state = 42)
```

## How to do train-test split for ML with pandas? (But with no overlap)

```python
train = ufo.sample(frac = 0.75, random_state = 42)
```

```python
test =  ufo.loc[~ufo.index.isin(train.index), :]  # no index of train overlaps in test
```
